### Visual Depiction of Ogive Geometries

Centering a sphere on a pipe (hollow cylinder) vs centering a cone in the same pipe. That's the **geometric effect** of hybrid/tangent vs secant ogive projectiles.

```python
import matplotlib.pyplot as plt
import numpy as np

# Diagram for ogive geometries
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Sphere in pipe (tangent ogive)
circle = plt.Circle((0, 0), 1, fill=False, color='blue', linewidth=2)
ax1.add_patch(circle)
ax1.plot([-1, 1], [0, 0], 'k-', linewidth=2)
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 2)
ax1.set_aspect('equal')
ax1.set_title('Tangent Ogive (Sphere-like)')
ax1.grid(True)

# Cone in pipe (secant ogive)
ax2.plot([-1, 0, 1], [0, 1, 0], 'r-', linewidth=2)
ax2.plot([-1, 1], [0, 0], 'k-', linewidth=2)
ax2.set_xlim(-2, 2)
ax2.set_ylim(-1, 2)
ax2.set_aspect('equal')
ax2.set_title('Secant Ogive (Cone-like)')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

# What We Really Mean by Consistency

## Defining Terms Honestly

Velocity consistency isn't one fluky low SD. Accuracy isn't single tight group.
Simply: Consistency means reliable results across lots of rounds. Not cherry-picking winners.
Population = every shot you could fire. Sample = handful you test. Samples mislead big time.

## Heart of Precision: Consistent Components

Most precision boils down to uniform ammo pieces.
Example: Unbalanced bullets wobble in flight. No handloader fix.

### Digging into Component Stats

We track what varies, why outcomes shift.
Tricks like uniform primer pockets depend on situation. Might not help. Could mess consistency.
Note: Top brands like Lapua, Alpha—brass weight, case volume don't always link.

### Picking Best Bullet for Job

Match bullet to task: hunting, target, long-range.
Hunting: focus on impact performance. Often speed-dependent.
Some bullets multi-role, depending on scenario.

### Ogive Shapes Explained Visually

Picture ball in pipe vs cone. That's tangent vs secant ogives. Sitting affects stability.

```python
import matplotlib.pyplot as plt
import numpy as np

# Diagram for ogive geometries
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Sphere in pipe (tangent ogive)
circle = plt.Circle((0, 0), 1, fill=False, color='blue', linewidth=2)
ax1.add_patch(circle)
ax1.plot([-1, 1], [0, 0], 'k-', linewidth=2)
ax1.set_xlim(-2, 2)
ax1.set_ylim(-2, 2)
ax1.set_aspect('equal')
ax1.set_title('Tangent Ogive (Sphere-like)')
ax1.grid(True)

# Cone in pipe (secant ogive)
ax2.plot([-1, 0, 1], [0, 1, 0], 'r-', linewidth=2)
ax2.plot([-1, 1], [0, 0], 'k-', linewidth=2)
ax2.set_xlim(-2, 2)
ax2.set_ylim(-1, 2)
ax2.set_aspect('equal')
ax2.set_title('Secant Ogive (Cone-like)')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

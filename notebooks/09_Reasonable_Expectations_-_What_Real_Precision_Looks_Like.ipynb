## Audette Ladders

Point of impact mostly from recoil. See the 400-yard demo.

```python
# WEZ simulation
import numpy as np
import matplotlib.pyplot as plt

# Basic hit odds
moa = 1
target_size = 12  # inches
range_yards = 400

# Rough calc
group_radius = (moa / 60) * range_yards * 36 / 12  # inches
if group_radius < target_size / 2:
    prob = 1
else:
    prob = (target_size / 2) / group_radius

print(f'Hit odds: {prob:.2f}')

# Wide plot
plt.figure(figsize=(10, 6))
plt.scatter([0], [0], s=group_radius*10)
plt.xlim(-target_size, target_size)
plt.ylim(-target_size, target_size)
plt.show()
```

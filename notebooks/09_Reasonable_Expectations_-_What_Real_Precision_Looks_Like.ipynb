## Audette Ladders

**POI** largely recoil induced. Demo of effect at 400y.

```python
# WEZ simulation
import numpy as np
import matplotlib.pyplot as plt

# Simple hit probability
moa = 1
target_size = 12  # inches
range_yards = 400

# Approximate
group_radius = (moa / 60) * range_yards * 36 / 12  # inches
if group_radius < target_size / 2:
    prob = 1
else:
    prob = (target_size / 2) / group_radius

print(f'Hit probability: {prob:.2f}')

# Plot
plt.scatter([0], [0], s=group_radius*10)
plt.xlim(-target_size, target_size)
plt.ylim(-target_size, target_size)
plt.show()
```

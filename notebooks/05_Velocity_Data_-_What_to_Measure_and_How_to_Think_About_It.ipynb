# Reading Velocity Data Right

## Making Sense of Chrono Numbers

Average speed matters for ballistics.
SD from few shots? Tricky. Fools you (Bramwell's 'perverse' quirk).
Smarter: Chart velocities in sequence for patterns. Check extreme spread on bigger batches. Track rolling SD.
'Flat spots' in ladders? Often noise faking steady zones.

```python
# Simulate velocity ladder
import numpy as np
import matplotlib.pyplot as plt

# Charge weights
charges = np.linspace(40, 50, 11)
# Velocities with some variation
velocities = 2600 + 20 * (charges - 45) + np.random.normal(0, 10, 11)

# Wide plot for mobile
plt.figure(figsize=(10, 6))
plt.plot(charges, velocities, 'o-')
plt.xlabel('Charge Weight (gr)')
plt.ylabel('Velocity (fps)')
plt.title('Velocity Ladder - Trends Over Flat Spots')
plt.show()
```

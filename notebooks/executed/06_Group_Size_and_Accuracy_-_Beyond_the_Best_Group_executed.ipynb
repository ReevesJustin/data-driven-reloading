# Group Size and Accuracy - Beyond the Best Group

## Goal: Teach honest on-target evaluation

Why **extreme spread** grows forever.
**Simple alternative:** Mean radius (average distance from center).
**Practical method:** Shoot 25–50 rounds at once on one big target (or aggregate multiple), measure mean radius.
**Interactive:** Overlay many 5-shot groups vs. one 50-shot aggregate from the same rifle.

## Precision vs. Accuracy

Define difference: **Precision** is how close shots are to each other; **accuracy** is how close to the target.
**Interactive demo:** Simulate scenarios where accuracy issues are mistaken for precision (e.g., wind shifts group center).

```python
# Simulate groups
import numpy as np
import matplotlib.pyplot as plt

# 5-shot groups
groups = [np.random.normal([0, 0], 0.5, (5, 2)) for _ in range(10)]

plt.figure(figsize=(8, 6))
for i, group in enumerate(groups):
    plt.scatter(group[:,0], group[:,1], label=f'Group {i+1}')
plt.legend()
plt.title('5-Shot Groups - Vary Due to Luck')
plt.show()

# Aggregate 50 shots
aggregate = np.random.normal([0, 0], 0.5, (50, 2))
plt.figure(figsize=(8, 6))
plt.scatter(aggregate[:,0], aggregate[:,1])
plt.title('50-Shot Aggregate - True Capability')
plt.show()
```

> **Key Takeaways**
> - Single group measurements are unreliable indicators of accuracy
> - Statistical analysis of group distributions reveals true capability
> - Outlier groups can skew perceptions of performance
> - Confidence intervals provide realistic accuracy expectations
> - Multiple group analysis prevents over-optimistic conclusions

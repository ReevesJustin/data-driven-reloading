# When IS a Result Real? — Power Analysis and Confidence Intervals

**Goal:** Introduce power analysis and confidence intervals for validating real data.

Content:

Explain statistical power: Probability of detecting a real effect.

Confidence intervals: Ranges where true value likely lies.

Interactive sims: Adjust sample size, effect size → see power curves.

Practical: "How many shots to be 80% sure your 5 fps difference is real?"

> **Key Takeaways**
> - Statistical significance testing determines if results are real
> - P-values and confidence levels quantify result reliability
> - Type I and Type II errors affect conclusion validity
> - Power analysis ensures adequate test sensitivity
> - Proper statistical methods validate or refute claims

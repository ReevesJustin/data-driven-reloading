# Group Size and Accuracy - Past the Best Group

## Fair Ways to Judge On-Target

Extreme spread just keeps climbing with more shots.
Try mean radius instead—average gap from the bull.
Real-world tip: Fire 25–50 rounds on a single large target (or combine groups), then measure that mean radius.
Interactive demo: Stack many 5-shot groups against one 50-shot bunch from the same gun.

## Precision Versus Accuracy

Precision means shots cluster tightly together; accuracy means they're near the aim point.
Demo: Sims showing how wind can fake precision problems as accuracy ones.

```python
# Simulate groups
import numpy as np
import matplotlib.pyplot as plt

# 5-shot groups
groups = [np.random.normal([0, 0], 0.5, (5, 2)) for _ in range(10)]

# Wide for mobile
plt.figure(figsize=(10, 6))
for i, group in enumerate(groups):
    plt.scatter(group[:,0], group[:,1], label=f'Group {i+1}')
plt.legend()
plt.title('5-Shot Groups - Luck Plays a Role')
plt.show()

# Aggregate 50 shots
aggregate = np.random.normal([0, 0], 0.5, (50, 2))
plt.figure(figsize=(10, 6))
plt.scatter(aggregate[:,0], aggregate[:,1])
plt.title('50-Shot Aggregate - Real Picture')
plt.show()
```

## Fair Ways to Judge On-Target

Try mean radius—avg gap from bull.

Demo: Stack 5-shot groups vs one 50-shot bunch from same gun.

## Precision vs Accuracy

Precision: shots cluster tight. Accuracy: near aim point.
Demo: Sims show wind faking precision as accuracy issues.

```python
# Simulate groups
import numpy as np
import matplotlib.pyplot as plt

# 5-shot groups
groups = [np.random.normal([0, 0], 0.5, (5, 2)) for _ in range(10)]

# Wide for mobile
plt.figure(figsize=(10, 6))
for i, group in enumerate(groups):
    plt.scatter(group[:,0], group[:,1], label=f'Group {i+1}')
plt.legend()
plt.title('5-Shot Groups - Luck Plays a Role')
plt.show()

# Aggregate 50 shots
aggregate = np.random.normal([0, 0], 0.5, (50, 2))
plt.figure(figsize=(10, 6))
plt.scatter(aggregate[:,0], aggregate[:,1])
plt.title('50-Shot Aggregate - Real Picture')
plt.show()
```

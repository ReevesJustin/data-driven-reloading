# Testing One Thing at a Time

## Goal: Teach simple, controlled experiments

**Common mistake:** Changing powder, charge, seating depth, and primer all at once.
**Rule:** Change one variable at a time, shoot enough rounds, then interpret carefully.
**Example walkthrough:** Testing two primers. Show proper setup, data collection template (CSV), basic plots.

```python
# Example data for primers
import pandas as pd
import matplotlib.pyplot as plt

# Simulated data
data = {
    'Primer': ['A'] * 30 + ['B'] * 30,
    'Velocity': np.random.normal(2700, 12, 30).tolist() + np.random.normal(2710, 12, 30).tolist()
}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 6))
for primer in ['A', 'B']:
    subset = df[df['Primer'] == primer]
    plt.hist(subset['Velocity'], alpha=0.5, label=f'Primer {primer}')
plt.legend()
plt.show()

# Stats
print(df.groupby('Primer')['Velocity'].describe())
```

> **Key Takeaways**
> - Multivariate testing can mask individual variable effects
> - Isolating variables ensures clear attribution of results
> - Sequential testing approach reveals true relationships
> - Confounding variables complicate interpretation
> - Systematic testing methodology leads to reliable conclusions

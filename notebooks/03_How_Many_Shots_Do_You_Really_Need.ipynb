# How Many Shots Do You Really Need?

## Practical Shot Counts

Let's simulate this: Say you're checking if one powder has less velocity spread than another—a true 10 fps vs. 15 fps gap. How many rounds per powder to spot it?
Try the slider below for confidence levels and shot needs.
Here are my everyday rules:
- Quick check: 20–30 shots (beats 10, but not foolproof).
- Solid test: 50–100 per tweak.
- Big claims like 'primer halves SD': 200+ shots.
Yeah, it's ammo-heavy. But trust me, it's cheaper than chasing mirages.

## Central Limit Theorem and SD Samples

Thanks to the central limit theorem, you need at least 30 shots to reliably gauge SD. If two 30-shot groups have close SDs, shoot more to confirm real differences.

## Shots for Precision Calls

To confidently say your rifle hits 1/2 MOA, you need 30 shots minimum.

## Averages Lock In Fast

Averages settle quickly. For zeroing or averaging velocity, 10 shots often nail it close enough.

```python
# Simulation for required shots
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from scipy import stats

# Function to simulate power for spotting difference
def simulate_power(n_shots, true_diff, sd):
    # Simulate two groups
    group1 = np.random.normal(2700, sd, n_shots)
    group2 = np.random.normal(2700 + true_diff, sd, n_shots)
    t_stat, p_value = stats.ttest_ind(group1, group2)
    return p_value < 0.05  # Significant at 5%

# Interactive slider
def plot_power(n):
    powers = [np.mean([simulate_power(i*10 + 10, 5, 12) for _ in range(100)]) for i in range(10)]
    # Wide for mobile
    plt.figure(figsize=(10, 6))
    plt.plot(range(10, 101, 10), powers)
    plt.xlabel('Sample Size')
    plt.ylabel('Power')
    plt.title(f'Power to Spot 5 fps Gap at n={n}')
    plt.show()

slider = widgets.IntSlider(value=30, min=10, max=100, step=10, description='Sample Size:')
widgets.interactive(plot_power, n=slider)
```
